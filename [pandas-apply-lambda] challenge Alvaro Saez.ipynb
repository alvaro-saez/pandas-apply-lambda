```python
def create_bin(data_set_votes):
    if data_set_votes >= 0 and data_set_votes < 1000:
        return 1
    elif data_set_votes >= 1000 and data_set_votes < 10000:
        return 2
    elif data_set_votes >= 10000 and data_set_votes < 100000:
        return 3
    elif data_set_votes >= 100000 and data_set_votes < 1000000:
        return 4
    elif data_set_votes >= 1000000:
        return 5
    else:
        return "error"
    
```

```python
bin = data_set["Votes"].apply(create_bin)
bin
```

CHALLENGE 4:
We want to know if the sum of the ASCII value of every character of the movie title divided by the number of votes retrieve a prime number...remember, prime numbers are integers

```python
def prime(title, votes):
    sum_title = 0
    for i in title:
        sum_title +=ord(i)
    if sum_title%votes == 0:
        return "prime"
    else:
        return "no prime"
```

```python
data_set["prime"] = data_set.apply(lambda data_set: prime(data_set['Title'], data_set['Votes']), axis=1)
data_set
```

We want to know which movies might have hidden paterns in their description. A way to know that is finding those movies which the sum of all numeric values of the string description hash (SHA256) are between their revenue and their number of votes.

```python
#hashlib.sha256()
def sha256(description, revenue, votes):
    enconde = description.encode()
    sha256 = hashlib.sha256(enconde).hexdigest()
    sha256_nums = re.sub("[a-zA-Z]","",sha256)
    sha256_nums_map = list(map(int, sha256_nums))
    sha256_nums_map_sum = sum(sha256_nums_map)
    if sha256_nums_map_sum < revenue*1000 and sha256_nums_map_sum > votes:
        return "hidden paterns"
    else:
        return "No hidden paterns"
```

```python
data_set["hidden paterns"] = data_set.apply(lambda data_set: sha256(data_set['Description'], data_set['Revenue (Millions)'], data_set['Votes']), axis=1)
data_set
```
